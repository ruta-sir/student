# Homework: Ultimate Battle

---

## Scenario

You are programming a game where 2 objects of your choosing (e.g., Robots, Dinosaurs, People) battle. Each object has health and power.

---

## Instructions

1. **Create a class** representing your object with:

   **Instance variables:**  
   - `String name`  
   - `int power`  
   - `int health`  
   - `MAKE YOUR OWN`

   **Static variables:**
   - `double fightDuration`

   **Instance methods:**  
   - `void attack()`
   - `void printStatus()`
   - `MAKE YOUR OWN`

   **Class methods:**  
   - `static int strongerFighter()`
   - `static void beginBattle()`
   - `MAKE YOUR OWN`

2. **In `main`:**  
   - Create two objects.  
   - Use instance methods to attack and print status.  
   - Use static methods to compare, print a fact, and start the battle.

```java
public class UltimateBattle {

    static class Robot {
        String name;
        int power;
        int health;
        String weapon; 

        static double fightDuration = 0;

        Robot(String name, int power, int health, String weapon) {
            this.name = name;
            this.power = power;
            this.health = health;
            this.weapon = weapon;
        }

        void attack(Robot enemy) {
            System.out.println(this.name + " attacks " + enemy.name + " with " + this.weapon + "!");
            enemy.health -= this.power;
            fightDuration += 1.5; // battle takes time
            if (enemy.health < 0) enemy.health = 0;
        }

        void printStatus() {
            System.out.println(name + " | Health: " + health + " | Power: " + power + " | Weapon: " + weapon);
        }

        void repair() {
            System.out.println(name + " is repairing itself...");
            health += 20;
            if (health > 100) health = 100;
        }

        static Robot strongerFighter(Robot r1, Robot r2) {
            if (r1.power > r2.power) return r1;
            else if (r2.power > r1.power) return r2;
            else return null; // equally strong
        }

        static void beginBattle(Robot r1, Robot r2) {
            System.out.println("\nThe Ultimate Battle Begins! ");
            System.out.println(r1.name + " vs " + r2.name + "!\n");

            while (r1.health > 0 && r2.health > 0) {
                r1.attack(r2);
                r2.printStatus();
                if (r2.health <= 0) break;

                r2.attack(r1);
                r1.printStatus();
            }

            System.out.println("\n Battle Over!");
            if (r1.health > 0) System.out.println(r1.name + " wins!");
            else if (r2.health > 0) System.out.println(r2.name + " wins!");
            else System.out.println("It's a draw!");

            System.out.println("Total fight duration: " + fightDuration + " seconds\n");
        }

        // Custom static method: fun fact
        static void printRobotFact() {
            System.out.println("Fun Fact: Robots never need sleep, but they do need recharging!");
        }
    }

    public static void main(String[] args) {
        // Create two robot objects
        Robot alpha = new Robot("AlphaX", 25, 100, "Laser Cannon");
        Robot beta = new Robot("BetaZ", 20, 100, "Plasma Blaster");

        alpha.printStatus();
        beta.printStatus();

        Robot stronger = Robot.strongerFighter(alpha, beta);
        if (stronger != null)
            System.out.println("\n" + stronger.name + " is the stronger fighter!");
        else
            System.out.println("\nBoth robots are equally powerful!");

        Robot.printRobotFact();

        Robot.beginBattle(alpha, beta);
    }
}
```
